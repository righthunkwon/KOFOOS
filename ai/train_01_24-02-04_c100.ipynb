# 학습
* class: 100
* epoch: 400
* optimizer: Adam
* batch size: 32
* 나머지 조정 없음

```python
#model = YOLO('yolov8m.pt')
model = YOLO("/home/j-i10a309/kofoos/train_01_24-02-04_c100_split_3/weights/last.pt")
```

```python
for k, v in model.named_parameters():
    print(k)
```

```python
results = {}

ksplit = 5
DATA_PATH = "/home/j-i10a309/datasets"
batch = 32
project = "kofoos"
basename = 'train_01_24-02-11_c100'
epochs = 400
optimizer="Adam"
patience=50
imgsz=640
lr0=1E-4
save_dir="./"
save_period=10
device='4'
```

```python
ds_yamls = []
for k in range(1, ksplit+1):
    basepath = os.path.join(DATA_PATH, f"split_{k}")
    ds_yamls.append(os.path.join(basepath, f"split_{k}_dataset.yaml"))
print(ds_yamls)
```

```python
for k in range(3, ksplit):
    dataset_yaml = ds_yamls[k]
    name = f"{basename}_split_{k+1}_"
    model.train(data=dataset_yaml, batch=batch, project=project, name=name, epochs=epochs, optimizer=optimizer, patience=patience, lr0=lr0, imgsz=imgsz, save_dir=save_dir, save_period=save_period, device=device)
    results[k] = model.metrics
```

```python
result = model.export(format='onnx')
```

# Test Dataset

```python
import os

DATA_PATH = "/home/j-i10a309/datasets/test"
```

```python
test_txt = os.path.join(DATA_PATH, "test.txt")
new_txt = os.path.join(DATA_PATH, "text_new.txt")
```

```python
# 1. D:/dataset/상품 이미지\images\val\ -> /home/j-i10a309/datasets/test/test/images/
# 2. \ -> /
with open(test_txt, "r", encoding="cp949") as original_file:
    with open(new_txt, "w", encoding="utf-8") as new_file:
        for line in original_file:
            # 특정 문자열 치환
            modified_line = line.replace("D:/dataset/상품 이미지\\images\\val\\", "/home/j-i10a309/datasets/test/test/images/")
            modified_line = modified_line.replace("\\", "/") 
            new_file.write(modified_line)
```
