```python
import pandas as pd
import xml.etree.ElementTree as ET
from pathlib import Path
import os
from tqdm import tqdm

DATA_PATH = 'D:\\dataset\\상품 이미지\\Validation\\annotations'
DATA_SAVE_PATH = Path('../csv_files')
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_row', None)
```

## 1. 제품 카테고리 만들기 

```python
# 큰 카테고리 폴더명 가져오기
dirs = [dir.name for dir in os.scandir(DATA_PATH) if not ".zip" in dir.name]
print(dirs)
```

```python
id = 0
category_dict = dict()
product_dict = dict()
all_dict = dict()
```

```python
for dir in dirs:
    new_path = os.path.join(DATA_PATH, dir)
    subdirs = [dir.name for dir in os.scandir(new_path) if not ".zip" in dir.name]
    for subdir in tqdm(subdirs):
        new_sub_path = os.path.join(new_path, subdir)
        file = [dir.name for dir in os.scandir(new_sub_path) if "meta.xml" in dir.name][0]

        category = []
        product = []
        all = []
        
        basename = os.path.basename(file)
        filename = os.path.splitext(basename)[0]
            
        tree = ET.parse(os.path.join(new_sub_path, file))
        root = tree.getroot().find("div_cd")
            
        # 카테고리 저장
        cat_1 = root.find("div_l").text
        cat_2 = root.find("div_m").text
        cat_3 = root.find("div_s").text
        cat_4 = root.find("div_n").text
        category.append(cat_1)
        category.append(cat_2)
        category.append(cat_3)
        category.append(cat_4)

        # 상품정보 저장
        item_no = root.find("item_no").text
        barcode = root.find("barcd").text
        name = root.find("img_prod_nm").text
        comp_nm = root.find("comp_nm").text
        product.append(item_no)
        product.append(barcode)
        product.append(name)
        product.append(comp_nm)


        all.append(item_no)
        all.append(barcode)
        all.append(name)
        all.append(comp_nm)
        all.append(cat_1)
        all.append(cat_2)
        all.append(cat_3)
        all.append(cat_4)


        category_dict[id] = category
        product_dict[id] = product
        all_dict[id] = all

        id += 1
```

```python
print(category_dict)
print(product_dict) # 6554
print(all_dict)
```

### 카테고리 정제

```python
all_df = pd.DataFrame.from_dict(data=all_dict, orient='index')
all_df.columns = ['item_no', 'barcode', 'name', 'comp_nm', "cat_1", "cat_2", "cat_3", "cat_4"]
all_df.head(5)
```

```python
before_keywords = pd.read_csv(os.path.join(DATA_SAVE_PATH, "before_keywords.csv"), encoding="utf-8")
new_keywords = pd.read_csv(os.path.join(DATA_SAVE_PATH, "new_keywords.csv"), encoding="utf-8")
```

```python
def replace_category(before, new):
    before_split = before.split(", ")
    new_split = new.split(", ")
    
    idxs = all_df[(all_df['cat_1']==before_split[0]) & (all_df['cat_2']==before_split[1]) & (all_df['cat_3']==before_split[2]) & (all_df['cat_4']==before_split[3])].index    
    for idx in idxs:
        all_df.iloc[idx, 4] = new_split[0]
        all_df.iloc[idx, 5] = new_split[1]
        all_df.iloc[idx, 6] = new_split[2]
        all_df.iloc[idx, 7] = new_split[3]
    #display(all_df.iloc[idxs])

for i in range(0, 314):
    replace_category(before=before_keywords.iloc[i].values[0], new=new_keywords.iloc[i].values[0])
```

```python
all_df.loc[all_df['cat_1'] == '면류', 'cat_1'] = '라면'
```

```python
category_df = all_df[['cat_1', 'cat_2','cat_3', 'cat_4']]
category_df = category_df.drop_duplicates(keep='first')
category_df = category_df.sort_values(by=['cat_1', 'cat_2', 'cat_3', 'cat_4']).reset_index(drop=True).reset_index()
category_df = category_df.rename(columns={"index":"id"})

print(f"{category_df.shape}")
category_df.head(5)
```

```python
category_df.to_csv(os.path.join(DATA_SAVE_PATH, "category.csv"), encoding="utf-8", index=False)
```

```python
result_df = pd.merge(category_df, all_df, left_on=['cat_1', 'cat_2', 'cat_3', 'cat_4'], right_on=['cat_1', 'cat_2', 'cat_3', 'cat_4'], how='inner')
result_df = result_df[['id', 'item_no', 'barcode', 'name', 'comp_nm']]
result_df = result_df.rename(columns={"id":"category_id"})
result_df = result_df.reset_index(drop=True).reset_index(drop=False).rename(columns={"index":"id"})
result_df.head(5)
```

```python
result_df.to_csv(os.path.join(DATA_SAVE_PATH, "products.csv"), encoding="utf-8", index=False)
```

```python
import pandas as pd
import xml.etree.ElementTree as ET
from pathlib import Path
import os
from tqdm import tqdm

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_row', None)
```

```python
DATA_PATH = "../csv_files"
product_df = pd.read_csv(os.path.join(DATA_PATH, "products.csv"), encoding="utf-8")
rayeon_df = pd.read_csv(os.path.join(DATA_PATH, "output.csv"), dtype={'report_no': str}, encoding="utf-8")
rayeon_df.head(3)
```

```python
product_merged_df = pd.merge(product_df, rayeon_df, on=["item_no"], how = "inner")
product_merged_df = product_merged_df.drop(columns=["barcode_y", "name_y", "comp_nm_y"])
product_merged_df = product_merged_df.rename(columns={"barcode_x":"barcode", "name_x":"name", "comp_nm_x":"comp_nm"})
```

```python
print(product_merged_df.shape)
product_merged_df.head(3)
```

```python
#product_merged_df = product_merged_df.drop(columns=["category_id", "item_no", "comp_nm"])
product_merged_df = product_merged_df.rename(columns = {"report_no":"prdlstReportNo", "id":"product_id"})
product_merged_df.head(3)
```

```python
product_to_disliked_df = pd.read_csv(os.path.join(DATA_PATH, "product_to_disliked.csv"), encoding="utf-8")
```

```python
product_to_disliked_df.head(2)
```

```python
tmp = product_to_disliked_df[["prdlstReportNo", "disliked_id"]]
tmp = tmp.drop_duplicates(keep='first')
tmp.head(3)
```

```python
product_to_disliked_new_df = pd.merge(product_merged_df, tmp, on="prdlstReportNo", how="inner")
product_to_disliked_new_df = product_to_disliked_new_df[["product_id", "name", "disliked_id", "item_no"]]
product_to_disliked_new_df = product_to_disliked_new_df.reset_index().reset_index(drop=True).rename(columns={"index":"id"})
print(product_to_disliked_new_df.shape)
```
